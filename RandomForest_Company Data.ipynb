```python
import pandas as pd
from sklearn.model_selection import KFold
##cross_val_score
from sklearn.model_selection import cross_val_score
##Label_Encoder-preprocessing
from sklearn.preprocessing import LabelEncoder
##StandardScalar-Preprocessing
from sklearn.preprocessing import StandardScaler
##RandomForestClassifier-ensemble method
from sklearn.ensemble import RandomForestClassifier
```

```python
data=pd.read_csv('Company_Data.csv')
data
```

```python
data['Sales'].median()
```

```python
data.describe
```

```python
data.loc[data['Sales']>=7,'sales']='High Sales'
data.loc[data['Sales']<7,'sales']='Low Sales'
data
```

```python
data_new=data.drop('Sales',axis=1)
data_new
```

```python
columns=['ShelveLoc','Urban','US','sales']
encoder=LabelEncoder()
for i in columns:
        data_new[i]=encoder.fit_transform(data_new[i])
data_new
```

```python
x=data_new.iloc[:,0:10]
y=data_new.iloc[:,10]
```

```python
max_features=4
max_estimators=100
kfold=KFold(n_splits=100,shuffle=True,random_state=42)
kfold
```

```python
Randomclf=RandomForestClassifier(n_estimators=100,max_features=max_features)
```

```python
Randomclf.fit(x,y)
```

```python
Randomclf.predict(x)
```

```python
cross_val=cross_val_score(Randomclf,x,y,cv=kfold)
cross_val
```
